```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow  #for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
#membuat list untuk menyimpan url dari beberapa image
urls = ["https://i.pinimg.com/736x/0f/47/2f/0f472f1719c1b8d3ff1f68a39e804453.jpg",
        "https://assets.kompasiana.com/items/album/2017/10/11/fa4eb39b-fcfe-47a2-b60f-d778a1297192-59dda141bde5757bb35671f5.jpeg?t=o&v=300",
        "https://www.rukita.co/stories/wp-content/uploads/2022/04/foto-kucing-oren.jpg"]

 #baca dan tampilkan image
 #loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                            #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)   #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  #convert color to rgb
  final_frame = cv.hconcat((image, image_2))         #concatenate image
  cv2_imshow(final_frame)                           #show image
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ", tinggi,"x", lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2), y] = [255,255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis vertikal ditengah image
for x in range (tinggi):
  image_3[x,int((lebar)/2)] = [255,255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis diagonal
for i in range(min(tinggi, lebar)):
    # Garis diagonal kiri atas ke kanan bawah
    image_3[i, i] = [255, 255, 255]

    # Garis diagonal kanan atas ke kiri bawah
    image_3[i, lebar - i - 1] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Tentukan posisi dan panjang garis
row = int(tinggi / 2)      # baris tengah
start_x = int(lebar / 4)
end_x = int(lebar / 4 * 3)

# Membuat garis horizontal putih di tengah dengan panjang tertentu
for y in range(start_x, end_x):
    image_3[row, y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# ukuran kotak
box_height = 50   # tinggi kotak dalam piksel
box_width = 50    # lebar kotak dalam piksel

# posisi kotak
start_row = 100
start_col = 150

# untuk mengatur kotak tidak keluar batas gambar
end_row = min(start_row + box_height, tinggi)
end_col = min(start_col + box_width, lebar)

for i in range(start_row, end_row):
    for j in range(start_col, end_col):
        image_3[i, j] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
